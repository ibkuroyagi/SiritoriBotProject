```python
import pandas as pd
```

```python
eki = pd.read_excel('wor_dict.xlsx',header=1)
eki = eki.dropna(subset=['読み'])
eki.head()
```

```python
# 使用するのはひらがなのみ
# 重複する文字はその重複回数分は使用してもよい
word_list = eki.groupby('読み')['読み'].count()
```

```python
word_dict = {}
for i in word_list.index:
    word_dict.update({i:word_list[i]})
word_dict
```

```python
w_list = list(pd.Series(word_list.index.values, word_list.values))
w_list
```

```python
import re

word = "りんご"
re_hiragana = re.compile(r'^[あ-ん]+$')
status_hira = re_hiragana.fullmatch(word) #fullmatch:完全一致
```

```python
if status_hira:
    print(word)
else:
    print(word*2)
```

```python
SMALL_TO_NORMAL = {
    'ぁ': 'あ', 'ぃ': 'い', 'ぅ': 'う', 'ぇ': 'え', 'ぉ': 'お',
    'っ': 'つ',
    'ゃ': 'や', 'ゅ': 'ゆ', 'ょ': 'よ',
}
def small2large(a="ぁ"):
    if a in SMALL_TO_NORMAL:
        return SMALL_TO_NORMAL[a]
    else:
        return a
```

```python
small2large("ょ")
```
